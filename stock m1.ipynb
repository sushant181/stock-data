# Stock Data

```python
import pandas as pd
import yfinance as yf
import datetime 
```

# list of Stocks

```python
df=pd.read_excel('G:\sushant\data A\List of Stocks.xlsx')
symbol=df.SYMBOL.to_list()
symbol
```

# remove the white space and concat the (.NS) for indian stock

```python
tickers=[i.strip() + '.NS' for i in symbol]
tickers
```

# featching the data 

```python
writer=pd.ExcelWriter('closeprice.xlsx')
for ticker in tickers:
    tickerdata=yf.Ticker(ticker)
    
    tickerdf=tickerdata.history(period='id', start='2020-6-4', end='2020-10-8')
    df=tickerdf.loc[:,['Close']]
    df.to_excel(writer,sheet_name=ticker)
    
writer.save()
```
